```python
import pandas as pd
```

```python
import matplotlib
```

```python
df = pd.read_csv('../data/raw/2019-05-28_portuguese_hate_speech_binary_classification.csv').reset_index(names="original_idx")
df.head()
```

```python
df = df.drop(df.columns[2:], axis=1)
df = df.rename(columns={})
```

```python
df.head()
```

```python
df.text = df.text.str.replace(r"https://\S*", "", regex=True)
df.text = df.text.str.replace(r"@\w*[:]?", "", regex=True)
df.text = df.text.str.replace(r" +", " ", regex=True)
df.text = df.text.str.strip("\n \t")
df.head()
```

```python
duplicated = df[df.text.duplicated()]
df = df[~df.original_idx.isin(duplicated.original_idx)]
duplicated.head()
```

```python
assert not (df.text.isna() | (df.text == "")).any()
assert not df.text.duplicated().any()
```
